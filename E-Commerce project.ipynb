```python
df.dropna(inplace=True)
```

```python
## Q1: Top 10 Revenue(sales) Generating Countries
top_sales = df.groupby('Country')['Sales'].sum().sort_values(ascending=False).head(10)
top_sales.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
## Q2: Top 3 Profitable Market
top_market = df.groupby('Market')['Profit'].sum().sort_values(ascending=False).head(3)
ax = sns.barplot(x=top_market.index, y=top_market.values, hue=top_market.index, palette='Blues_r')
plt.xlabel('Market')
plt.ylabel('Profit')
plt.title('Top 3 Profitable Market')
plt.legend(title="Market", labels=top_market.index, loc='upper right')
plt.xticks(rotation=0)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
## Q3: Top 10 Subcategory for profit Ratio
top_subcategory = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)
top_subcategory.plot(kind='pie',autopct='%1.1f%%')
plt.title('Top 10 Sub-Category for Profit Ratio')
plt.show()
```

```python
## Q4: Bottom 5 Revenue(Sales) generating Sub-Category  
Bottom_subcategory = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).tail(5)
Bottom_subcategory.plot(kind='line', marker="o", color="blue", linewidth=2)
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.title('Bottom 5 Revenue(Sales) generating Sub-Category')
plt.xticks(rotation=0)
plt.grid(axis='both', linestyle='dotted')
plt.show()
```

```python
## Q5: Revanue(Sales) per E-Commerce platform
ab= df.groupby('E-Commerce Site')['Sales'].sum().sort_values(ascending=False)
c = ['red','blue','orange']
ax = ab.plot(kind='bar',width=0.5, color=c)
plt.xlabel('E-Commerce ')
plt.ylabel('Sales')
plt.title('Revenue(Sales) per E-Commerce platform')
plt.xticks(rotation=0)
plt.show()
```

```python
## Q6: Order priority-wise Quantity
order = df.groupby('Order Priority')['Quantity'].count().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
ax = sns.scatterplot(x=order.index, y=order.values, color='red', s=200)
for i, txt in enumerate(order.values):
    plt.text(order.index[i], order.values[i], str(txt), ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.xlabel('Order Priority')
plt.ylabel('Quantity')
plt.title('Order Priority-wise Quantity (Scatter Plot)')
plt.xticks(rotation=0)
plt.show()
```

```python
## Q7: Revenue(Sales) Growth per year
sales_growth = df.groupby(df['Ship Date'].dt.year)['Sales'].sum()
sns.lineplot(x=sales_growth.index, y=sales_growth.values, marker="o", color="blue", linewidth=2)
for i, txt in enumerate(sales_growth.values):
    plt.text(sales_growth.index[i], sales_growth.values[i], f'{txt:,.0f}',  
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Sales Growth Per Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
## Q8: Total Revenue(Sales)
total_revenue = df['Sales'].sum()
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=["Total Revenue"], y=[total_revenue], color="lightblue")
plt.text(0, total_revenue, f"${total_revenue:,.2f}", ha='center', va='bottom', fontsize=12, fontweight='bold')
plt.ylabel('Revenue ($)')
plt.title('Total Revenue (Sales)')
plt.ylim(0, total_revenue * 1.1)
plt.show()
```

```python
## Q9: Total Profit
total_Profit = df['Profit'].sum()
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=["Total Profit"], y=[total_Profit], color="lightblue")
plt.text(0, total_Profit, f"${total_Profit:,.2f}", ha='center', va='bottom', fontsize=12, fontweight='bold')
plt.ylabel('Profit ($)')
plt.title('Total Profit')
plt.ylim(0, total_Profit * 1.1)
plt.show()
```

```python
## Q10: NUmber of product Got Sold per Year
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
products_sold_per_year = df.groupby('Year')['Quantity'].sum()
print(products_sold_per_year)
plt.figure(figsize=(8, 5))
plt.bar(products_sold_per_year.index, products_sold_per_year.values, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Products Sold')
plt.title('Number of Products Sold Per Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
